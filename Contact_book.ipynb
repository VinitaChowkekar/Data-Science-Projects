```python
contacts={}
def add_contact():
  name=input("Enter name: ")
  contact=int(input("Enter the number: "))
  contacts[name]=contact
  print("Contact added successfully")
  print("-----------------------------------------")
def search_contact():
  name=input("Enter the name to search: ")
  if name in contacts:
    print("Contact found")
    print("Name: ",name)
    print("Contact Number: ",contacts[name])
    print("-----------------------------------------")
  else:
    print("Contact not found")
    print("-----------------------------------------")
def update_contact():
  name=input("Enter the name to update: ")
  if name in contacts:
    contact=int(input("Enter the new number: "))
    contacts[name]=contact
    print("Contact updated successfully")
    print("-----------------------------------------")
  else:
    print("Contact not found")
    print("-----------------------------------------")
def delete_contact():
  name=input("Enter the name to delete: ")
  if name in contacts:
    del contacts[name]
    print("Contact deleted successfully")
    print("-----------------------------------------")
  else:
    print("Contact not found")
def all_contacts():
  for name,contact in contacts.items():
    print("Name: ",name)
    print("Contact Number: ",contact)
    print("-----------------------------------------")
print("Welcome to Contact Book")
print("1. Add Contact")
print("2. Search Contact")
print("3. Update Contact")
print("4. Delete Contact")
print("5. All contacts")
print("6. Exit")
ch=input("Enter the number of operation you want to perform: ")
if ch == '1':
  add_contact()
elif ch == '2':
  search_contact()
elif ch == '3':
  update_contact()
elif ch == '4':
  delete_contact()
elif ch == '5':
  all_contacts()
elif ch == '6':
  print("Contact Book closed successfully")
  exit()
else:
  print("Invalid operation, Enter the valid operation.")
```
