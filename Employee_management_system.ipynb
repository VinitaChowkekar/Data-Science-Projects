```python
employees = {}  # Dictionary to store employee data

def add_employee():
  """Adds a new employee to the system."""
  name = input("Enter employee name: ")
  id = int(input("Enter employee ID: "))
  designation = input("Enter employee position: ")
  salary = float(input("Enter employee salary: "))

  if id in employees:
    print("Employee ID already exists!")
  else:
    employees[id] = {
      "name": name,
      "designation": designation,
      "salary": salary
    }
    print("Employee added successfully!")
    print("-" * 20)

def remove_employee():
  """Removes an employee from the system."""
  id = int(input("Enter employee ID to remove: "))
  if id in employees:
    del employees[id]
    print("Employee removed successfully!")
    print("-" * 20)
  else:
    print("Employee not found!")
    print("-" * 20)

def display_employees():
  """Displays information of all employees."""
  if not employees:
    print("No employees found!")
  else:
    print("\nEmployee Details:")
    for id, employee in employees.items():
      print(f"\nID: {id}")
      print(f"Name: {employee['name']}")
      print(f"Designation: {employee['designation']}")
      print(f"Salary: {employee['salary']}")
      print("-" * 20)

def EMS():
  print("\nEmployee Management System")
  """Main function to run the employee management system."""
  while True:
    print("\n1. Add Employee")
    print("2. Remove Employee")
    print("3. Display Employees")
    print("4. Exit")

    choice = input("Enter your choice: ")
    print("-" * 20)
    print("Results:")

    if choice == '1':
      add_employee()
    elif choice == '2':
      remove_employee()
    elif choice == '3':
      display_employees()
    elif choice == '4':
      print("Exiting the program. Goodbye!")
      print("-" * 20)
      break
    else:
      print("Invalid choice! Please try again.")
      print("-" * 20)

if __name__ == "_main_":
  main()

EMS()
```
